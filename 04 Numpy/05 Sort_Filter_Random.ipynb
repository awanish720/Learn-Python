## Sorting numpy array

```python
import numpy as np

arr = np.array([7,6,2,0,1])
print(np.sort(arr))
```

```python
arr = np.array(["orange","banana","apple","cherry"])
print(np.sort(arr))
```

```python
arr = np.array([True,False,True,False,True])
print(np.sort(arr))
```

```python
## 2-D Array
arr = np.array([[6,3,0],[1,9,4]])
print(np.sort(arr))
```

```python
## 3-D Array
arr = np.array([[[2,3,0]],[[9,2,1]]])
print(np.sort(arr))
```

### Filter the array

```python
## Filter the array
arr = np.array([23,67,1,89,56])
x = [False,True,False,True,True]

newarr = arr[x]
print(newarr)
```

```python
filter_arr = arr > 50
newarr = arr[filter_arr]
print(filter_arr)
print(newarr)
```

```python
filter_arr = arr%2 == 1
newarr = arr[filter_arr]
print(filter_arr)
print(newarr)
```

## Rondom Number in numpy

```python
import numpy as np

print(np.random.randint(0,101))
```

```python
from numpy import random

x = random.randint(100)
print(x)
```

```python
x = random.rand()
print(x)
```

```python
# Generating random 1-D array
x = random.randint(150,size=(6))
print(x)
```

```python
# 2-D array random 
x = random.randint(200, size=(4,5))
print(x)
```

```python
# 3-D array random
x = random.randint(51,size=(3,4,5))
print(x)
```

```python
# float value
x = random.rand(10)
print(x)
```

```python
x = random.rand(2,4)
print(x)
```

```python
x = random.rand(2,3,4)
print(x)
```

```python
# Choice()
# 1-D array
x = random.choice([3,6,2,5,9])
print(x)
```

```python
# 2-D array
x = random.choice([2,5,7,4,8,9], size=(3,4))
print(x)
```

```python
# 3-D array
x = random.choice([3,5,7,1,8],size=(2,3,4))
print(x)
```

```python
# choice() with probability
x = random.choice([2,4,7,9],size=100, p=[0.1,0.3,0.6,0.0])
print(x)
```

```python
# 2-D Array
x = random.choice([3,5,2,7],size=(3,4),p=[0.3,0.5,0.0,0.2])
print(x)
```

## Random Permutations with numpy

```python
# shuffle()
arr = [2,4,6,8,3]
random.shuffle(arr)
print(arr)
```

```python
# permutation()
arr = [2,4,6,8,3]
x = random.permutation(arr)

print("Original: ",arr)
print("Shuffled Copy: ",x)
```
