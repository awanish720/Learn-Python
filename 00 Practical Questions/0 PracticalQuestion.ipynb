3. Write a Python program that calculates compound interest using the formula A = P(1 + r/n)^(nt), where P, r, n, and t are user inputs.

```python
P = float(input("Enter principal: "))
r = float(input("Enter rate: "))
n = int(input("Enter times compounded per year: "))
t = int(input("Enter year: "))

A = P * (1 + r/n)**(n*t)
print("Principal Amount: ",P)
print("Rate: ",r)
print("Times compounded per year: ",n)
print("Year: ",t)
print(f"Amount: Rs {A:.2f}")
```

```python
# Write a Python program to create a set of vowels ("a", "e", "i", "o", "u"). 
# Check if each letter in a word (e.g.,"hello") is in the vowel set and print the vowels found ?

vowels = {"a","e","i","o","u"}

word = "Python"

found_vowels = [char for char in word if char in vowels]
print("Vowels Found: ",found_vowels)
```

```python
#Write a Python program to create a dictionary with 3 students and their test scores. Calculate the average
#score and print the student with the highest score ?

students = {"Akash": 85,"Divya": 92,"Chetan": 78}

average_score = sum(students.values()) / len(students)
print("Average score: ",average_score)

highest_scorer = max(students, key=students.get)
print("Students with highest score: ",highest_scorer, "with", students[highest_scorer])
```

```python
# Create a program that opens a file specified by the user and reads its contents. Handle the FileNotFoundError to display an error message if the file does not exist.
def read_file():
    filename = input("Enter filename: ")
    try:
        with open(filename, 'r') as file:
            contents = file.read()
            print("Python Program")
            print(contents)
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found!")

read_file()
```

```python
# 5. Write a function that attempts to access a dictionary key provided by the user. Handle the KeyError and return a default value (e.g., "Key not found") if the key does not exist.

def dict_key(dict, key):
    try:
        return dict[key]
    except KeyError:
        return "Key not Found"
    
if __name__ == "__main__":
    print("-- Dictionary Key Access--")
    sample_dict = {"name":"Ayush","age":25,"city":"Chennai"}
    key = input("Enter dictionary key to access: ")
    result = dict_key(sample_dict, key)
    print(f"Result: {result}")
```
