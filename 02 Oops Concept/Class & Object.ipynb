## Class and Objects

```python
# define a class
class Dog:
    sound = "bark" # class attribute

dog1 = Dog() # object of the class

print(dog1.sound) # access the class attribute
```

```python
# Using --init__()
class Dog:
    species = "Canine" 

    def __init__(self, name, age):
        self.name = name # Instance attribute
        self.age = age

dog1 = Dog("Buddy",3)

print(dog1.name)
print(dog1.age)
print(dog1.species)
```

```python
# Self parameter
class Dog:
    def __init__(self, name, age):
        self.name = name # Instance attribute
        self.age = age
    
    def bark(self):
        print(f"{self.name} is barking!")

dog1 = Dog("Husky",4)
dog1.bark()
```

```python
# __str_method
class Dog:
    def __init__(self, name, age):
        self.name = name # Instance attribute
        self.age = age
    
    def __str__(self):
        return f"{self.name} is {self.age} year old."
    
dog1 = Dog("Charlie",5)
dog2 = Dog("Buddy",3)

print(dog1)
print(dog2)
```

### Class and Instance Variables in Python

```python
class Dog:
    species = "Canine" 

    def __init__(self, name, age):
        self.name = name # Instance attribute
        self.age = age

dog1 = Dog("Buddy", 3)
dog2 = Dog("Charlie",5)

print(dog1.species) # class variable
print(dog1.name) # Instance variable
print(dog2.age) 

dog1.name = "Max"
print(dog1.name) # update instance variable

Dog.species = "Feline"
print(dog1.species) # updated class variable
print(dog2.species)
```

```python
class Car:
    wheels = 4

    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def display_info(self):
        print(f"{self.year} {self.brand} {self.model} with {self.wheels} wheels")

    def drive(self):
        print(f"The {self.brand} {self.model} is driving!")

car1 = Car("Toyota","Camry",2020)
car2 = Car("Honda","Civic",2022)

print(car1.brand)
print(car2.year)
print(car1.wheels)

car1.display_info()
car2.drive()
```
