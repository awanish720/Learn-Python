# Tuples in Python

Tuple Operations:
1. Creating tuples
2. Accessing elements
3. Length of tuple
4. Concatenation 
5. Repetition
6. Tuple Methods
7. Nested tuple

```python
# Creating tuples
empty_tuple = ()  # Empty tuple
single_tuple = (42,)  # Single-element tuple (note the comma)
mixed_tuple = (1, "hello", 3.14, True)  # Mixed data types
nested_tuple = (1, (2, 3), 4)
print(empty_tuple)
print(single_tuple)
print(mixed_tuple)
print(nested_tuple)
```

```python
# Accessing elements
print("First element of mixed_tuple:", mixed_tuple[0])  
print("Last element of mixed_tuple:", mixed_tuple[-1])  
print("Slice [1:3]:", mixed_tuple[1:3])
```

```python
# Length of tuple
print("Length of mixed_tuple:", len(mixed_tuple))
```

```python
# Membership testing
print("Is 'hello' in mixed_tuple?:", "hello" in mixed_tuple)
```

```python
# Concatenation
tuple1 = (1, 2)
tuple2 = (3, 4)
combined = tuple1 + tuple2
print("Concatenated tuple:", combined)
```

```python
# Repetition
repeated = tuple1 * 2
print("Repeated tuple:", repeated)
```

```python
# Unpacking a tuple
x, y, z, w = mixed_tuple # mixed_tuple = (1, "hello", 3.14, True) 
print(f"Unpacked values: x={x}, y={y}, z={z}, w={w}")
```

```python
# Tuple methods
print("Count of 1 in mixed_tuple:", mixed_tuple.count(1))
print("Index of 'hello' in mixed_tuple:", mixed_tuple.index("hello"))
```

```python
# Nested tuple access
# nested_tuple = (1, (2, 3), 4)
print("Nested tuple second element:", nested_tuple[1][0])
```
