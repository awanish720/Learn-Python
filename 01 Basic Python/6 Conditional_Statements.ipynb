```python
# If statement
age = 20
if age >= 18:
    print("You are eligible to vote.")
```

```python
# If-else statement
num = 5
if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
# if-elif-else statement
marks = 85
if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

```python
# Nested if-else statement
x = 10
if x > 0:
    if x % 2 == 0:
        print("Positive even number")
    else:
        print("Positive odd number")
```

```python
# Nested if-else
age = 50
is_member = False
if age >= 60:
    if is_member:
        print("30% senior discount!")
    else:
        print("20% senior discount.")
else:
    print("Not eligible for a senior discount.")
```

```python
x = 10
y = 20
result = "Both are Positive" if x > 0 and y > 0 else "At least one is non-positive"
print(result) 
```

```python
score = 65
grade = "A" if score >=90 else "B" if score >=80 else "C"
print(grade)
```

```python
# Ternary Conditional Statement in Python
# A ternary conditional statement is a compact way to write an if-else condition in a single line. 
# Itâ€™s sometimes called a "conditional expression."

# Syntax: value_if_true if condition else value_if_false

age = 16
status = "Adult" if age >= 18 else "Minor"
print(status)
```

```python
num = 7
result = "Even" if num % 2 == 0 else "Odd"
print(result)
```

```python
a = 10
b = 20
max_value = a if a > b else b
print("Maximum is:", max_value)
```
