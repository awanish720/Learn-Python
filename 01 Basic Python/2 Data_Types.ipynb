```python
# Numeric - int
x = 10
print("Value of x:", x)
print("Type of x:", type(x))
```

```python
# Numeric - float
y = 3.14
print("Value of y:", y)
print("Type of y:", type(y))
```

```python
# Sequence - str

s = "Python"
print("String:", s)
print("First character:", s[0])
```

```python
# Sequence - list

l = [1, 2, 3]
print("List: ",l)
l.append(4)
print("New List:", l)
```

```python
# Sequence - tuple

t = (1, 2, 3)
print("Tuple:", t)
print("Second item:", t[1])
```

```python
# Sequence - range

r = range(5)
print("Range values:")
for i in r:
    print(i, end=" ")
```

```python
# Set - set

s = {1, 2, 3, 2}
print("Set (no duplicates):", s)
```

```python
# Mapping - dict

person = {"name": "Awanish", "age": 25}
print("Name:", person["name"])
print("Age:", person["age"])
```

```python
# Boolean - bool

is_passed = True
print("Passed:", is_passed)
print("Type:", type(is_passed))
```
