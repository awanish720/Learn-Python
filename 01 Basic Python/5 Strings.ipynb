```python
# Single and double quotes
name = "Abhishek"
greeting = 'Hello, World!'
print(name)
print(greeting)  

# Multiline string with triple quotes
multiline = """This is a
multiline
string."""
print(multiline)
```

```python
# String are Array

a = "Python Programming"
print(a[1])
```

```python
# String Slicing

b = "Python is High level Language."
print(b[2:5])

# Slicing from start
print(b[:6])

# Slicing from end
print(b[4:])

# Negative Indexing
print(b[-5:-2])
```

```python
b = "Python is High level Language."
split_b = b.split('')
print(split_b)
```

```python
# String Concatination

str1 = "Hello"
str2 = "World"
result = str1 + str2
print("Concatenated String:", result)
```

```python
# String Repetition

str1 = "Python"
result = str1 * 3
print("Repeated String:", result)
```

```python
# Indexing

str1 = "Python Programming"
char = str1[7]  
print("Character at index 1:", char)
```

```python
# Length of the String

str1 = "Python Programming"
length = len(str1)
print("Length of the string:", length)
```

```python
# Common String Methods 

# Define a string
text = " Hello, Python! "

# Using upper() to convert to uppercase
uppercase_text = text.upper()
print("Uppercase:",uppercase_text)

# Using lower() to convert to lowercase
lowercase_text = text.lower()
print("Lowercase:",lowercase_text)

# Using strip() to remove whitespace
stripped_text = text.strip()
print("Stripped:",stripped_text)

# Using replace() to replace a substring
replaced_text = text.replace("Hello","Hi")
print("Replaced:",replaced_text)

# Using split() to split a string into a list
split_text = text.split(",")
print("Split:",split_text)

# Using join() to join elements of a list into a string
joined_text = "-".join(["a","b","c"])
print("Joined:",joined_text)

for i in text:
    print(i)
```

```python
# String Format

# Using f-strings (Python 3.6+)
name = "RAJ"
age = 25
message = f"My name is {name} and I am {age} years old."
print(message)

# Using .format() method
message = "My name is {} and I am {} years old.".format(name, age)
print(message)
```
