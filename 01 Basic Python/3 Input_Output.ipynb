Input Operation

```python
# Taking User Input with input()
# Simple string input

name = input("Enter your name: ")
print("Hello,", name)
```

```python
# Converting Input to Integer with int()
# Taking numeric input and using it in a calculation

age = int(input("Enter your age: "))
print("Next year, you'll be", age + 1)
```

```python
# Converting Input to Float with float()
# Accepting float input

height = float(input("Enter your height in meters: "))
print("Your height in centimeters is:", height * 100)
```

```python
# Taking Multiple Inputs
# Accepting multiple inputs separately
name = input("Enter your name: ")
city = input("Enter your city: ")
print(name, "lives in", city)
```

Output Operation

```python
# Basic Console Output using print()

print("Hello, World!")

# Using Multiple Items in print()

print("Akash", 25, "Python")
```

```python
# Using sep to Change Separator

print("Akash", 25, "Python", sep=", ")
```

```python
# Using end to Control Line End

print("Processing...", end="\n")
print("Done!")
```

```python
# Using print() with Lists or Variables

languages = ["Python", "Java", "C++"]
print("Popular languages:", languages)
```

```python
# Formatted Print

name = "Bharti"
age = 30
print(f"{name} is {age} years old.")
```

```python
# calculator

a = int(input("Enter First number: "))
b = int(input("Enter second number: "))

add = a+b
sub = a-b
mult = a*b
div = a/b
floor_div = a//b
mod = a%b
exp = a**b

print("Addition = ",add)
print("Subtraction = ",sub)
print("Multiplication = ",mult)
print("Division = ",div)
print("Floor Division = ",floor_div)
print("Modulus = ",mod)
print("Exponentiation = ",exp)
```
