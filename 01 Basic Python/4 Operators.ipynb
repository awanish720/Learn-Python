2. Comparison Operators

```python
## Comparison Operators

# equal to 
a = 5
b = 5
result = a == b
print("Is a equal to b?", result)
```

```python
# Not equal 

a = 5
b = 3
result = a != b
print("Is a not equal to b?", result)
```

```python
# greater than > 
a = 5
b = 2
result = a > b
print("Is a greater than b?", result)
```

```python
# Less than < 

a = 5
b = 3
result = a < b
print("Is a less than b?", result)
```

```python
# Greater Than or Equal To (>=)

a = 5
b = 3
result = a >= b
print("Is a greater than or equal to b?", result)
```

```python
# Less Than or Equal To (<=)

a = 5
b = 3
result = a <= b
print("Is a less than or equal to b?", result)
```

3. Logical Operators

```python
# Logical AND

a = True
b = False
result = a and b
print("True and False =", result)
```

```python
# Logical OR

a = True
b = False
result = a or b
print("True or False =", result)
```

```python
# Logical NOT (not)

a = True
result = not a
print("not True =", result)
```

```python
# Use case
x = 5
y = 10

print("x > 3 and y < 15:", x > 3 and y < 15)  
print("x < 3 or y < 5:", x < 3 or y < 5)      
print("not(x > y):", not(x > y))
```

4. Identity Operators

```python
# is

a = [1, 2, 3]
b = a
print("a is b:", a is b)
```

```python
# is not

a = [1, 2, 3]
b = [1, 2, 3]
print("a is not b:", a is not b)
```

```python
# Check Identity with id() Function

x = 10
y = 10
print("x is y:", x is y)
print("id(x):", id(x))
print("id(y):", id(y))
```

5. Membership Operators

```python
# in

numbers = [1, 2, 3]
result = 4 in numbers
print("Is 3 in the list?", result)
```

```python
# not in 

numbers = [1, 2, 3]
result = 4 not in numbers
print("Is 4 not in the list?", result)
```

6. Bitwise Operators

```python
# Bitwise AND &    1 & 1 => 1

a = 5      # 0101 in binary       0101 => 1010
b = 3      # 0011 in binary       0011
result = a & b                  # 0001
                                # 0111
                                # 0110
print("5 & 3 =", result)
```

```python
# Bitwise OR (|)  0 or 0 => 0

a = 5
b = 3
result = a | b
print("5 | 3 =", result)
```

```python
# Bitwise XOR (^) => 0 ^ 0 => 0 , 1 ^ 1 => 0

a = 5
b = 3
result = a ^ b
print("5 ^ 3 =", result)
```

```python
# Bitwise NOT (~)

a = 5
result = ~a
print("~5 =", result)
```

```python
# Left Shift (<<)

a = 5           # 5 =  0101  => 
result = a << 1
print("5 << 1 =", result)
```

```python
# Right Shift (>>)

a = 5  
result = a >> 1
print("5 >> 1 =", result)
```

7. Assignment Operators

```python
# Assignment (=)
x = 10
print("x =", x)
```

```python
# Add and assign
x = 10
x += 3  # same as x = x + 3
print("x =", x)
```

```python
# subtract and assign
x = 10
x -= 3  # same as x = x - 3
print("x =", x)
```

```python
# Multiply and assign
x = 10
x *= 3  # same as x = x * 3
print("x =", x)
```

```python
# Divide and Assign
x = 10
x /= 3  # same as x = x / 3
print("x =", x)
```

```python
# Floor Divide and Assign
x = 10
x //= 3  # same as x = x // 3
print("x =", x)
```

```python
# Modulus and Assign
x = 10
x %= 3  # same as x = x % 3
print("x =", x)
```

```python
# Exponent and Assign
x = 10
x **= 3  # same as x = x ** 3
print("x =", x)
```
