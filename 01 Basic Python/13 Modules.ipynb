# Modules in Python

```python
# Using a Built-in Modules
import math

# Using math module functions
print(math.sqrt(16))  
print(math.pi)        
print(math.factorial(5))  
```

```python
# Using a Standard Library Module: Random
import random

# Generate a random integer between 1 and 10
print(random.randint(1, 10))  
# Choose a random item from a list
fruits = ['apple', 'banana', 'orange']
print(random.choice(fruits))
```

```python
# Creating and Using a Custom Module 
# main.py
import mymodule

# Use functions and variables from mymodule
print(mymodule.greet("Akash"))  
print(mymodule.pi)              
print(mymodule.square(5))
```

```python
# Importing Specific Items
from mymodule import greet, pi

print(greet("Bharti"))  
print(pi)           
```

```python
# Using an Alias
import mymodule as mm

print(mm.greet("Chandan"))  
print(mm.square(3))        
```

```python
# Using the os Module
import os

# Get current working directory
print(os.getcwd())
# List files in a directory
print(os.listdir('.'))  
```

```python
# Using the datetime Module
from datetime import datetime

# Get current date and time
now = datetime.now()
print(now)  
# Format the date
print(now.strftime("%Y-%m-%d")) 
```
