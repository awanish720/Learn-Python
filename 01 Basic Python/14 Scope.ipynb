### Python Scope

```python
# Local variable
def calculate_total():
    price = 500  # Local variable
    print(f"Total price: {price}")

calculate_total()
```

```python
# Function Inside Function
def outer_calculation():
    total = 1000  # Enclosing variable
    
    def inner_calculation():
        print(f"Inner total: {total}")  # Accesses enclosing variable
    
    inner_calculation()

outer_calculation()
```

```python
# Global Scope
count = 50  # Global variable

def display_count():
    print(f"Inside function: {count}")  # Accesses global variable

display_count()
print(f"Outside function: {count}")  # Accesses global variable
```

```python
# Naming Variables
score = 100  # Global variable

def update_score():
    score = 50  # Local variable, shadows global score
    print(f"Inside function: {score}")

update_score()
print(f"Outside function: {score}")
```

```python
# Global Keyword

def set_value():
    global counter  # Declare counter as global
    counter = 500  # Assign value to global variable

set_value()
print(f"Global counter: {counter}")
```

```python
# Nonlocal Keyword
def outer_task():
    status = "pending"  # Enclosing variable
    
    def inner_task():
        nonlocal status  # Declare status as nonlocal to modify enclosing variable
        status = "completed"  # Modify enclosing status
    
    inner_task()
    return status

print(outer_task())
```
