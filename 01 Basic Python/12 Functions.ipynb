# Functions in Python

```python
## Why Function?
num=24
if num % 2 == 0:
    print("the number is even")
else:
    print("the number is odd")
```

```python
# Simple Function without Parameters
def greet():
    print("Hello, welcome to Python!")

# Calling the function
greet()
```

```python
# Function with Parameters
def add_numbers(a, b):
    sum = a + b
    return sum

# Calling the function
result = add_numbers(5, 3)
print("Sum:", result)
```

```python
# Function with Default Parameters
def introduce(name, greeting="Hello"):
    return f"{greeting}, {name}!"

# Calling the function
print(introduce("Alice"))  # Uses default greeting
print(introduce("Bob", "Hi"))  # Custom greeting
```

```python
## Variable Length Arguments
## Positional and Keywords arguments

# Positional Arguments
def print_numbers(*args):
    for number in args:
        print(number)

print_numbers(1,2,3,4,5,6,7,8,"Abhi")
```

```python
## Keywords Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="Abhi",age="25",country="India")
```

```python
# Function with *args and **kwargs

def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional agrument: {val}")

    for key,value in kwargs.items():
        print(f"{key}: {value}")

print_details(1,2,3,4,"Abhi",name="Awanish",age="27",country="India")
```

```python
## Return Statements
def multiply(a,b):
    return a*b

multiply(2,3)
```

```python
# Function With Multiple Return Values

def divide_and_remainder(a, b):
    quotient = a // b
    remainder = a % b
    return quotient, remainder

# Calling the function
q, r = divide_and_remainder(10, 3)
print("Quotient:", q, "Remainder:", r)
```

## Recursive Function

Recursion is a method that breaks a problem into smaller sub-problems and calls itself for each of them.

```python
# Factorial of number

def factorial(n):
    if n == 0 or n == 1:  # Base case
        return 1
    return n * factorial(n - 1)  # Recursive case

# Calling the function
print("Factorial of 5:", factorial(5))
```

```python
# Fibonacci Sequence is where each number is the sum of the two preceding ones: 0, 1, 1, 2, 3, 5, 8, ....

def fibonacci(n):
    if n <= 1:  # Base case
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)  # Recursive case

# Calling the function
print("Fibonacci of 6:", fibonacci(6))
```

## Lambda Function

A lambda function in Python is a small, anonymous function defined using the lambda keyword.

```python
# add number
x = lambda a : a + 10
print(x(7))
```

```python
# multiply number
x = lambda a, b : a * b
print(x(8, 6))
```

```python
# Lambda function to square a number
square = lambda x: x * x

# Calling the lambda function
# result = square(8)
print("Square of 8:", square(8))
```

```python
# within a function
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)

print(mydoubler(11))
```

```python
## Lambda with map()

 # List of temperatures in Celsius
celsius = [0, 20, 37, 100]

# Lambda to convert Celsius to Fahrenheit
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Celsius to Fahrenheit:", fahrenheit)
```

```python
## Lambda with filter()

# List of numbers
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

# Lambda to filter even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))

print("Even numbers:", evens)
```

```python
## Lambda with sorted()

# List of tuples (name, score)
students = [("Aarav", 85), ("Bhavya", 92), ("Charan", 78), ("Divya", 95)]

# Sort by score using lambda
sorted_students = sorted(students, key=lambda x: x[1])

print("Sorted by score:", sorted_students)
```

```python
## Lambda with Multiple Arguments

# Lambda to calculate hypotenuse (sqrt(a^2 + b^2))
hypotenuse = lambda a, b: (a**2 + b**2)**0.5

# Example usage
a, b = 3, 4
result = hypotenuse(a, b)
print(f"Hypotenuse of triangle with sides {a} and {b}: {result:.2f}")
```

## Nested Function

```python
## Basic Nested Function
def outer_function(a, b):
    def inner_function(x):
        return x * x
    total = a + b
    return inner_function(total)

# Calling the function
result = outer_function(3, 4)
print("Square of sum (3+4):", result)
```

```python
## Closure with Nested Function
def make_multiplier(factor):
    def multiplier(number):
        return number * factor
    return multiplier

# Create multiplier functions
double = make_multiplier(2)
triple = make_multiplier(3)

# Use the multiplier functions
print("Double of 5:", double(5))
print("Triple of 5:", triple(5))
```
