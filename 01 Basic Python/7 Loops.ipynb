Python has two types of loops:

1. for loop

2. while loop


# For Loop

```python
# range(stop) function 
for i in range(5):
    print(i)
```

```python
# range(start,stop) function
for i in range(2, 6):
    print(i)
```

```python
# range(start,stop,step) function
for i in range(1, 10, 2):
    print(i)
```

```python
# python program to find sum of number from 1 to 5
total = 0
for i in range(1, 6):
    total += i
print("Sum:", total)
```

```python
# Using range() in Reverse Order
for i in range(10, 0,-1):
    print(i)
```

```python
# Using range() in Reverse Order
for i in range(10, 0, -2):
    print(i)
```

```python
# for loop in list operation
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

```python
# Looping Through a String
for x in "banana":
  print(x)
```

```python
# Else in For Loop
for x in range(6):
  print(x)
else:
  print("Finally finished!")
```

# While Loop

```python
# count 1 to 5
count = 1
while count <= 5:
    print("Count:", count)
    count += 1
```

```python
# python program to print even num from 1 to 10
i = 2
while i <= 10:
    print(i)
    i += 2
```

```python
# Else in while loop
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```

```python
# python program to print num from 1 to 10 in reverse order
i = 10
while i >= 1:
    print(i)
    i -= 1
```

```python
# Infinite While Loop
while True:
    user_input = input("Enter 'q' to quit: ")
    if user_input == 'q':
        break
    print("You entered:", user_input)
```

# Nested Loops

### Nested For Loop

```python
# Printing a 3x3 Grid of Stars
for row in range(3):
    for col in range(3):
        print("*", end=" ")
    print()
```

```python
# Multiplication Table (1 to 5)
for i in range(1, 6):
    for j in range(1, 6):
        print(i * j, end="\t")
    print()
```

### Nested While Loop

```python
# Simple Grid with Nested while
i = 1
while i <= 3:
    j = 1
    while j <= 3:
        print("*", end=" ")
        j += 1
    print()
    i += 1
```

```python
# Number Pattern
i = 1
while i <= 5:
    j = 1
    while j <= i:
        print(j, end=" ")
        j += 1
    print()
    i += 1
```

# Loop control statements
Break

Continue

Pass


```python
# break
for i in range(1, 6):
    if i == 3:
        break
    print(i)
```

```python
# continue
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

```python
# pass
for i in range(1, 6):
    if i == 3:
        pass  # Reserved for future logic
    print(i)
```
