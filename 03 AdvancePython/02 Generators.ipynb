## Generators in Python


```python
def fun(max):
    cnt = 1
    while cnt <= max:
        yield cnt
        cnt += 1

ctr = fun(5)
for n in ctr:
    print(n)
```

```python
def fun():
    yield 1
    yield 2
    yield 3

for val in fun():
    print(val)
```

```python
# return example
def fun():
    return 1 + 2 + 3

res = fun()
print(res)
```

```python
# Generator Expression
sq = (x*x for x in range(1,11)) # for x in range(1,11):
                                    # sq = x*x
                                    # print(sq)
for i in sq:
    print(i)
```
