## Bar Chart

```python
import matplotlib.pyplot as plt

fruits = ['Apples', 'Bananas','Cherries','Dates']
sales = [400,350,300,450]

plt.bar(fruits, sales)
plt.title('Fruits Sales')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.show()
```

```python
# customizing bar chart
colors = ['red', 'yellow', 'orange', 'black']

plt.bar(fruits, sales, color=colors, width=0.5)
plt.title('Fruits Sales')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.show()
```

```python
# creating horizontal bar
colors = ['red', 'yellow', 'orange', 'black']

plt.barh(fruits, sales, color=colors)
plt.title('Fruits Sales')
plt.xlabel('Sales')
plt.ylabel('Fruits')
plt.show()
```

```python
# Multiple bar plots
import numpy as np

barWidth = 0.25
fig = plt.subplots(figsize=(12,8))

IT = [12,30,1,8,22]
ECE = [28,6,16,5,10]
CSE = [29,3,25,23,17]

br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

plt.bar(br1, IT, color='r', width= barWidth, edgecolor = 'black', label='IT')
plt.bar(br2, ECE, color='g', width= barWidth, edgecolor = 'black', label='ECE')
plt.bar(br3, CSE, color='b', width= barWidth, edgecolor = 'black', label='CSE')

plt.title("Multiple Bar Plots")
plt.xlabel('Branch', fontweight='bold', fontsize = 15)
plt.ylabel('Students Passed', fontweight='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(IT))], ['2015','2016','2017','2018','2019'])
plt.legend()
plt.show()
```

```python
# Stacked bar Plot
import numpy as np
N = 5 # no. of bar

boys = np.array([20,35,30,35,27])
girls = np.array([25,32,34,20,25])
boyStd = (3,4,5,7,9)
girlStd = (3,5,4,6,9)
ind = np.arange(N)
width = 0.35

# percent calculations
totals = boys + girls
percent_boys = (boys/totals)*100
percent_girls = (girls / totals)* 100

fig, ax = plt.subplots(figsize=(10,7))
p1 = plt.bar(ind, boys, width, yerr = boyStd)
p2 = plt.bar(ind, girls, width, bottom=boys, yerr= girlStd)

# add labels
for i in range(N):
    ax.text(ind[i], boys[i]/2, f'{percent_boys[i]:.1f}%' ,ha='center',va='center',color='white', fontsize=10, fontweight='bold')

    ax.text(ind[i], boys[i] + girls[i]/2, f'{percent_girls[i]:.1f}%' ,ha='center',va='center',color='white', fontsize=10, fontweight='bold')

plt.title('Contribution by the Teams')
plt.xlabel('Teams')
plt.ylabel('Contribution')
plt.xticks(ind, ('T1','T2','T3','T4','T5'))
plt.yticks(np.arange(0,81,10))
plt.legend((p1[0],p2[0]),('boys','girls'))
plt.show()
```
