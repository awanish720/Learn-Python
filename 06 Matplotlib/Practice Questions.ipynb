### Create a line chart to visualize stock prices over a week using Matplotlib. Use the data: dates = ['2025-06-16', '2025-06-17', '2025-06-18', '2025-06-19','2025-06-20'] and prices = [100, 102, 99, 105, 108]. Ensure the x-axis displays dates properly formatted.

```python
# Line Chart for Stack Prices
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

dates = ['2025-06-16', '2025-06-17', '2025-06-18', '2025-06-19','2025-06-20']
prices = [100, 102, 99, 105, 108]

date_objects = [datetime.strptime(date, '%Y-%m-%d') for date in dates]

plt.figure(figsize=(10,5))
plt.plot(date_objects, prices, marker='o', linestyle='-', color='blue')

# Format the X-axis
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())

plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Stock Prices Over a week')
plt.grid(True)

plt.tight_layout()
plt.show()
```
