## Line Chart

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = [20,25,35,55]

plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.show()
```

```python
# simple line
import numpy as np

x = np.array([1,2,3,4])
y = x*2

plt.plot(x,y)
plt.title("Simple Line Chart")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
# Customizing Line Chart

x = [10,20,30,40]
y = [20,25,35,55]

plt.plot(x,y, color='red', linewidth=3, marker='o', markersize=10, linestyle='--')

plt.title("Customizing Line Chart")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
# Line chart with Annotations
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.figure(figsize=(8,8))
plt.plot(x,y, marker='o', linestyle='-')

# add annotations
for i, (xi,yi) in enumerate(zip(x,y)):
    plt.annotate(f'({xi},{yi})', (xi,yi), textcoords="offset points", xytext=(0,10), ha='center')

plt.title("Line Chart with Annotations")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.grid(True)
plt.show()
```

```python
# Multiple Line Chart
x = np.array([1,2,3,4])
y = x*2

plt.plot(x,y)
plt.title("MultiLine Plot")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()

plt.figure()
x1 = [2,4,6,8]
y1 = [3,5,7,9]

plt.plot(x1, y1, '-.')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
# Subplot
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [2,4,6,8,10]

plt.figure(figsize=(9,5))

plt.subplot(3,2,1)
plt.plot(x,y1, color='green')
plt.title("Plot 1")

plt.subplot(3,2,2)
plt.plot(y1,x, color='red')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y2, color='blue')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(y2,x, color='violet')
plt.title("Plot 4")
plt.show()
```

```python
# Fill the Area Between two Lines 
x = np.array([1,2,3,4])
y = x*2

plt.plot(x,y)
plt.show()

x1 = [2,4,6,8]
y1 = [3,5,7,9]

plt.plot(x,y1, '-.')
plt.show()

plt.plot(x,y1, '-.')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Multiple Plots")

plt.fill_between(x,y,y1, color='red', alpha=0.5)
plt.show()
```
