## Concatenation DataFrame

```python
# 1. using .concat()
import pandas as pd

data1 = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
         'Age': [27, 24, 22, 32],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Msc', 'MA', 'MCA', 'Phd']}

data2 = {'Name': ['Abhi', 'Ayushi', 'Dhiraj', 'Hitesh'],
         'Age': [17, 14, 12, 52],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Btech', 'B.A', 'Bcom', 'B.hons']}

df = pd.DataFrame(data1, index=[0,1,2,3])
df1 = pd.DataFrame(data2, index=[4,5,6,7])
print(df, "\n\n", df1)
```

```python
frames = [df,df1]
res1 = pd.concat(frames)
res1
```

```python
# 2. Concat by setting logic on Axes

data1 = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
         'Age': [27, 24, 22, 32],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Msc', 'MA', 'MCA', 'Phd'],
         'Mobile No': [97,91,58,76]}

data2 = {'Name': ['Abhi', 'Ayushi', 'Dhiraj', 'Hitesh'],
         'Age': [17, 14, 12, 52],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Btech', 'B.A', 'Bcom', 'B.hons'],
         'Salary': [10000,20000,30000,40000]}

df = pd.DataFrame(data1, index=[0,1,2,3])
df1 = pd.DataFrame(data2, index=[2,3,6,7])

print(df, "\n\n", df1)
```

```python
# inner join => intersection
res2 = pd.concat([df,df1], axis=1, join='inner')
res2
```

```python
# outer join => Union
res3 = pd.concat([df,df1], axis=1, join='outer')
res3
```

```python
# 3. Concat Df by Ignoring Indexes
res = pd.concat([df,df1], ignore_index=True)
res 
```

```python
# 4. Concat df with Group keys
frames = [df,df1]
res = pd.concat(frames, keys=['X','Y'])
res
```

```python
# 5. Concat Mixed df and series
data1 = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
         'Age': [27, 24, 22, 32],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Msc', 'MA', 'MCA', 'Phd']}

df = pd.DataFrame(data1, index=[0,1,2,3])

s1 = pd.Series([10000,20000,30000,40000], name='Salary')

print(df, "\n\n", s1)
```

```python
result = pd.concat([df,s1], axis=1)
result
```

## Merging DataFrame

```python
# 1. Merging using One Key
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],}

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']}

df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)

print(df, "\n\n", df1)
```

```python
res = pd.merge(df, df1, on='key')
res
```

```python
# 2. Merging using Multiple keys

data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K1', 'K0', 'K1'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],}

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K0', 'K0', 'K0'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']}

df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)

print(df, "\n\n", df1)
```

```python
res1 = pd.merge(df, df1, on=['key','key1'])
res1 
```

```python
# 3. Merging df using How argument
res = pd.merge(df,df1, how='left', on=['key','key1'])
res
```

```python
res = pd.merge(df,df1, how='right', on=['key','key1'])
res
```

```python
res = pd.merge(df,df1, how='outer', on=['key','key1'])
res
```

```python
res = pd.merge(df,df1, how='inner', on=['key','key1'])
res
```

## Joining DataFrame

```python
# 1. Using .join()

data1 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32]}

data2 = {'Address':['Allahabad', 'Kannuaj', 'Allahabad', 'Kannuaj'],
        'Qualification':['MCA', 'Phd', 'Bcom', 'B.hons']}

df = pd.DataFrame(data1,index=['K0', 'K1', 'K2', 'K3'])

df1 = pd.DataFrame(data2, index=['K0', 'K2', 'K3', 'K4'])

print(df, "\n\n" , df1)
```

```python
res = df.join(df1)
res
```

```python
# 2. joining using "on" Argument
data1 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],
        'Key': ['K0', 'K1', 'K2', 'K3']}

data2 = {'Address':['Allahabad', 'Kannuaj', 'Allahabad', 'Kannuaj'],
        'Qualification':['MCA', 'Phd', 'Bcom', 'B.hons']}
df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2, index=['K0', 'K2','K3', 'K4'])

res2 = df.join(df1, on='Key')
res2
```

```python
# 3. Join with different Index Levels(Multi-Index)

data1 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32]}

data2 = {'Address':['Allahabad', 'Kannuaj', 'Allahabad', 'Kannuaj'],
        'Qualification':['MCA', 'Phd', 'Bcom', 'B.hons']}

df = pd.DataFrame(data1, index=pd.Index(['K0','K1','K2','K3'], name='key'))

index = pd.MultiIndex.from_tuples([('K0','Y0'), ('K1', 'Y1'), ('K2','Y2'), ('K2','Y3')], names=['key', 'Y'])

df1 = pd.DataFrame(data2, index= index)

print(df, "\n\n", df1)
```

```python
result = df.join(df1, how='inner')
result
```
