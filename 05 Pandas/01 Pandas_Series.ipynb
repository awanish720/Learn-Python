## Pandas Series

```python
# Creating an Empty Series
import pandas as pd

ser = pd.Series()
print(ser)
```

```python
import numpy as np
import pandas as pd

data = np.array(['P','y','t','h','o','n'])

ser = pd.Series(data)
print(ser)
```

```python
# from the list

data_list = [23,5,6,12,7]

ser = pd.Series(data_list)
print(ser)
```

```python
# from a dictionary
data_dict = {'Name':'Raj','Age':25,'Gender':'Male'}
ser = pd.Series(data_dict)
print(ser)
```

```python
# using range()

ser = pd.Series(range(2,20))
print(ser)
```

```python
# using list comprehension

ser = pd.Series(range(1,20,3), index=[x for x in 'abcdefg'])
print(ser)
```

```python
# Accessing elements

a = [3,5,7,9,4]
ser = pd.Series(a)
print(ser)
```

```python
# using labels
print(ser[2])
print(ser[4])
```

```python
# create labels
ser = pd.Series(a, index=["A","B","C","D","E"])
print(ser)
```

### Binary Operations on Pandas Series

```python
# 1. Arithmetic operations

s1 = pd.Series([10,20,30], index=['a','b','c'])
s2 = pd.Series([2,4,6], index=['a','b','c'])

# add 
add = s1 + s2
mult = s1 * s2
sub = s1 - s2
div = s1 / s2

print(add)
print(mult)
print(sub)
print(div)
```

```python
# 2. Comparison Operations

s1 = pd.Series([11,34,67])
s2 = pd.Series([11,45,34])

equal = s1 == s2
greater = s1 > s2
less = s1 < s2

print(equal)
print(greater)
print(less)
```

```python
# 3. Logical AND on Series
s1 = pd.Series([True,False,True])
s2 = pd.Series([False,False,True])

res = s1 & s2
print(res)
```

### Pandas Series Index Attribute

```python
data = pd.Series([10,20,30,40], index=['a','b','c','d'])

print("Original Index: ",data.index)

# modify the index
data.index= ['W','X','Y','Z']
print("Modified Series: \n",data)
```

```python
# change the specific index with rename()
data = pd.Series([10,20,30,40], index=['a','b','c','d'])
data1 = data.rename(index={'b':'X'})
print("Modified Series: \n",data1)
```

```python
# assinging duplicate index labels

ser = pd.Series(['Mumbai','Delhi','Chennai','Bangalore','Bihar'], index= ['City 1','City 1', 'City 3', 'City 3', 'City 2'])
print(ser)
```
