```python
# Handling Nested Json in pandas
import json
from pandas import json_normalize

sample_json = {
    "organizations": [
        {
            "org_id": 1,
            "name": "Bharatiya Kala Kendra",
            "city": "Delhi",
            "events": [
                {"event_id": 101, "title": "Kathak Dance Recital", "date": "2025-07-15", "duration_hours": 2},
                {"event_id": 102, "title": "Hindustani Classical Concert", "date": "2025-08-10", "duration_hours": 3}
            ]
        },
        {
            "org_id": 2,
            "name": "Nritya Academy",
            "city": "Mumbai",
            "events": [
                {"event_id": 201, "title": "Bharatanatyam Performance", "date": "2025-06-20", "duration_hours": 2.5},
                {"event_id": 202, "title": "Odissi Workshop", "date": "2025-09-05", "duration_hours": 4}
            ]
        },
        {
            "org_id": 3,
            "name": "Sangeet Natak Akademi",
            "city": "Chennai",
            "events": [
                {"event_id": 301, "title": "Carnatic Music Festival", "date": "2025-10-12", "duration_hours": 5}
            ]
        }
    ]
}

# save to file
with open("indian_cultural_events.json",'w') as f:
    json.dump(sample_json, f, indent=2)

# load the Json File
with open('indian_cultural_events.json','r') as f:
    d = json.load(f)

# Normalize the 'events' data
events_df = json_normalize(d['organizations'], record_path='events', meta=['org_id','name','city'])
events_df
```

```python
# JSON Orientations in pandas
data = {
    'event_id': [101,102,201],
    'title': ['Kathak Recital', 'Classical Concert', 'Bharatanatyam Show'],
    'duration_hours': [2.0, 3.0,2.5]
}

df = pd.DataFrame(data, index=['E1','E2','E3'])
print("Original DataFrame: ")
df
```
