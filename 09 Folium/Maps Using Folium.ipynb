```python
df['Lat'] = ""
df['Long'] = ""
for index in df.index:
    df.loc[df.index==index, 'Lat']= locations[index][0]
    df.loc[df.index==index, 'Long']= locations[index][1]

df
```

```python
India = folium.Map(location=[23,80], tiles='cartodbpositron', zoom_start=5,max_zoom=7, height=500, width='80%')
India
```

```python
for i in range(0, len(df.index)):
    folium.Circle(
        location=[df.iloc[i]['Lat'],df.iloc[i]['Long']],
        tooltip= df.iloc[i].name+
            '<li>Confirmed: '+str(df.iloc[i]['Confirmed'])+'</li>'+
            '<li>Recovered: '+str(df.iloc[i]['Cured'])+'</li>'+
            '<li>Deaths: '+str(df.iloc[i]['Deaths'])+'</li>',
        radius=int((df.iloc[i]['Confirmed'])*.03), 
        color = "#f77721" 
    ).add_to(India)
India
```
