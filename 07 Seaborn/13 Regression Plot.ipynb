## Regression Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.regplot(x='total_bill',y='tip', data=tips, scatter_kws={'s':10}, line_kws={'color':'red'})
plt.title("Regression Plot b/w Total Bill vs Tips")
plt.show()
```

```python
data = sns.load_dataset('mpg')
data
```

```python
sns.regplot(x='mpg', y='acceleration', data=data)
plt.title("Regression Plot b/w MPG vs Accelerations")
plt.show()
```

```python
# Polynomial Regression (non-linear)
sns.regplot(x='total_bill', y='tip', data=tips, order=2)
plt.title("Polynomial Regression (Degree 2)")
plt.show()
```

```python
# Separate Regression Lines by Category
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips)
plt.title('Tip Vs Total Bill by Sex')
plt.show()
```

```python
# Regression with Logistic Function
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.regplot(x='age', y='survived', data=titanic, logistic=True, ci=99)
plt.title('Logistic Regression: Age vs Survival')
plt.show()
```

```python
# Faceted Regression by Multiple Categories
sns.lmplot(x='total_bill', y='tip', col='sex', row='smoker', data=tips)
plt.show()
```

```python
# Lowess Regression (Locally Weighted Smoothing)
sns.regplot(x='total_bill', y='tip', data=tips, lowess=True)
plt.title("Lowess Regression")
plt.show()
```

```python
# Regression with Marker Style and Line Customization
sns.regplot(x='total_bill', y='tip', data=tips, marker='x', color='green', line_kws={'color':'red', "linewidth": 2})
plt.title("Customized Regression Line and Markers")
plt.show()
```

```python
# Strip Plot + Regression Overlay
sns.stripplot(x='sex', y='tip', data=tips, jitter=True)
sns.pointplot(x='sex', y='tip', data=tips, color='red', linestyles='none')
plt.title("Categorical Strip Plot with Regression Points")
plt.show()
```
