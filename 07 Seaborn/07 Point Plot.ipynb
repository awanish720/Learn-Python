## Point Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'Name':['RAVI','MEENA','TARUN','PRIYA'],
        'Age':[22,25,23,26]}
df = pd.DataFrame(data)

sns.pointplot(x='Name', y='Age', data=df)
plt.title("Point Plot b/w Name and Age")
plt.show()
```

```python
data = sns.load_dataset('tips')
sns.pointplot(x='sex', y='total_bill', data=data, hue='smoker')
plt.title("Total Bill Vs Sex")
plt.show()
```

```python
sns.pointplot(x='size', y='total_bill', data=data, hue='sex', linestyles='-.', markers='^')
plt.title("Total Bill Vs Size")
plt.show()
```
