## KDE Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris_df = sns.load_dataset('iris')
sns.kdeplot(data=iris_df[iris_df['species'] == 'virginica'],
            x='sepal_length', color='b',fill=True, label='Iris_Virginica')
plt.ylabel('Probability Density')
plt.title("KDE Plot of Iris data")
plt.legend()
plt.show()
```

```python
sns.kdeplot(data=iris_df[iris_df['species'] == 'virginica'],
            x='sepal_length', color='b',fill=True, label='Iris_Virginica')

sns.kdeplot(data=iris_df[iris_df['species'] == 'setosa'],
            x='sepal_length', color='r',fill=True, label='Iris_Setosa')
plt.title("KDE Plot of Sepal Length")
plt.xlabel('Sepal Length')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
# 2D KDE Plot
iris_setosa = iris_df.query("species == 'setosa'") # query(expr, inplace=False )
iris_virginica = iris_df.query("species == 'virginica'")

sns.kdeplot(x=iris_setosa['sepal_length'], y=iris_setosa['sepal_width'], cmap='Reds', fill= True, label='Iris_Setosa')

sns.kdeplot(x=iris_virginica['sepal_length'], y=iris_virginica['sepal_width'], cmap='Blues', fill= True, label='Iris_Virginica')

plt.title("2D KDE Plot: Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.show()
```
