## Scatter Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'Name': ['RAVI','MEENA','TARUN','PRIYA'],
        'Age':[22,25,23,26]
        }

df = pd.DataFrame(data)

sns.set_style('whitegrid')
sns.scatterplot(x=df.index, y='Age', data=df)
plt.title("Scatter Plot")
plt.show()
```

```python
# Using tips dataset
tips = sns.load_dataset('tips')

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter Plot of Total Bill vs Tips")
plt.show()
```

```python
sns.scatterplot(x='day', y='tip', data=tips)
plt.title("Scatter Plot of Day vs Tips")
plt.show()
```

```python
# using fmri
sns.load_dataset('fmri')
```

```python
sns.set_style('whitegrid')
fmri = sns.load_dataset('fmri')

sns.scatterplot(x='timepoint', y='signal', data=fmri)
plt.title("FMRI Signal over Time")
plt.show()
```

```python
sns.scatterplot(x='timepoint', y='signal', data=fmri, hue='region', style='event')
plt.title("FMRI Signal by Region and Event")
plt.show()
```

```python
# iris dataset
sns.scatterplot(x='sepal_length', y='sepal_width', data=sns.load_dataset('iris'))
plt.title('Sepal Legth vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
sns.scatterplot(x='petal_length', y='petal_width', data=sns.load_dataset('iris'))
plt.title('Petal Legth vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```
