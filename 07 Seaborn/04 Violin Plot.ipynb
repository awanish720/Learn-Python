## Violin Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'Name': ['Ansh','Sahil', 'Jayan', 'Anurag'],
        'Age':[21,23,20,24]}
df = pd.DataFrame(data)

sns.violinplot(y='Age', data=df)
plt.title("Violin Plot of Age")
plt.show()
```

```python
# Using Tips data 
sns.set_style('whitegrid')

tips = sns.load_dataset('tips')
plt.figure(figsize=(6,7))

sns.violinplot(x='day', y='tip', data=tips, hue='day', inner='quartile', palette=["#FF5500CB", "#E8D319C9", "#15FF00CA","#00AAFFCA"])
plt.title("Violin Plot Tips Vs Day")
plt.show()
```

```python
# Using Iris Dataset
custom_palette = sns.color_palette("husl", 8)
sns.set_palette(custom_palette)
data = sns.load_dataset('iris')
sns.violinplot(x='species', y='petal_length', data=data)
plt.title('Petal Length Distribution by Species')
plt.show()
```

```python
# swarmplot()
sns.violinplot(x='day',y='tip', data=tips, hue='day', palette=["#FF5500CB", "#E8D319C9", "#15FF00CA","#00AAFFCA"])
sns.swarmplot(x='day', y='tip',data=tips, color='White', size=3)
plt.title("Violin Plot with Swarmplot b/w Day vs Tip")
plt.show()
```
