```python
import pandas as pd
import seaborn as sns

data ={'Name':['AYUSH','SAHIL','JAYAN','ANURAG'], 'Age':[21,23,20,24]}
df = pd.DataFrame(data)

sns.lineplot(x=df.index, y='Age', data=df, color='red')
df
```

```python
sns.load_dataset('flights')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

sns.lineplot(x='year', y='passengers', data=sns.load_dataset('flights'))

plt.title('Number of Passengers Over Time')
plt.show()
```

```python
# Adding Markers to Line Plots

sns.lineplot(x='year', y='passengers', data=sns.load_dataset('flights'), marker='o', color='green')
plt.title('Number of Passengers Over Time')
plt.show()
```
