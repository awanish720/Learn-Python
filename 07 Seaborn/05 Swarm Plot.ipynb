## Swarm Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')
sns.swarmplot(x='day', y='total_bill', data=data, size=5, hue='day', palette=['#FF6F61',"#61B5FF","#B061FF","#BAFF61"])
plt.title('Swarm Plot of Total Bill vs Day')
plt.show()
```

```python
# Oriented to horizontal
sns.swarmplot(y='day', x='total_bill', data=data, size=5, hue='size', palette=['#FF6F61',"#61B5FF","#B061FF","#BAFF61"], orient='h')
plt.title('Swarm Plot of Total Bill vs Day')
plt.show()
```

```python
sns.swarmplot(x='sex', y='total_bill', data=data, size=5, hue='smoker', dodge=True)
plt.title('Swarm Plot of Total Bill vs Day')
plt.show()
```
