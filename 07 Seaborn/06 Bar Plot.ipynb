## Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')
df
```

```python
sns.barplot(x='class', y='fare', data=df, hue='class')
plt.title("Bar Plot Class vs Fare")
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex')
plt.title("Bar Plot Class vs Fare")
plt.show()
```

```python
sns.barplot(y='class', x='fare', data=df, hue='sex')
plt.title("Bar Plot Class vs Fare")
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, hue='class', order=["Third", "Second", "First"])
plt.title("Bar Plot Class vs Fare")
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex',palette='husl')
plt.title("Bar Plot Class vs Fare")
plt.show()
```
