## Box Plot in Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'Name': ['Ansh','Sahil', 'Jayan', 'Anurag'],
        'Age':[21,23,20,24]}
df = pd.DataFrame(data)

sns.boxplot(y='Age', data=df)
plt.title("Box Plot by Age")
plt.show()
```

```python
tips = sns.load_dataset('tips')
sns.boxplot(x='day',y='total_bill', data=tips, color='green')
plt.title("Box Plot Total Bill by Day")
plt.show()
```

```python
sns.set_style('whitegrid')
sns.boxplot(x='species',y='petal_length', data=sns.load_dataset('iris'),hue='species')
plt.title('Petal Length Distribution by Species')
plt.show()
```
